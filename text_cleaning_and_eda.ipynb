```python
import re
import string
import numpy as np
import pandas as pd
from tqdm.auto import tqdm
import nltk
from nltk.stem import WordNetLemmatizer
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk import word_tokenize

import matplotlib.pyplot as plt
import plotly.express as px
from plotly.graph_objs import *
```

```python
df = pd.read_json('training_json_file.json')
```

```python
filter = df['cleaned_html'] != ''
filtered_df = df[filter]
filtered_df = filtered_df.dropna()
```

```python
# Some Standard Text Cleaning Functions

def convert_to_lower(text):
    return text.lower()

def remove_numbers(text):
    number_pattern = r'\d+'
    without_number = re.sub(pattern=number_pattern, repl=" ", string=text)
    return without_number

def remove_punctuation(text):
    return text.translate(str.maketrans('', '', string.punctuation))

def remove_stopwords(text):
    removed = []
    stop_words = list(stopwords.words("english"))
    tokens = word_tokenize(text)
    for i in range(len(tokens)):
        if tokens[i] not in stop_words:
            removed.append(tokens[i])
    return " ".join(removed)

def remove_extra_white_spaces(text):
    single_char_pattern = r'\s+[a-zA-Z]\s+'
    without_sc = re.sub(pattern=single_char_pattern, repl=" ", string=text)
    return without_sc
```

```python
filtered_df['cleaned_html'] = tqdm(filtered_df['cleaned_html'].apply(lambda x: convert_to_lower(x)))
filtered_df['cleaned_html'] = tqdm(filtered_df['cleaned_html'].apply(lambda x: remove_numbers(x)))
filtered_df['cleaned_html'] = tqdm(filtered_df['cleaned_html'].apply(lambda x: remove_punctuation(x)))
filtered_df['cleaned_html'] = tqdm(filtered_df['cleaned_html'].apply(lambda x: remove_stopwords(x)))
filtered_df['cleaned_html'] = tqdm(filtered_df['cleaned_html'].apply(lambda x: remove_extra_white_spaces(x)))
```

```python
filtered_df['length_text'] = filtered_df['cleaned_html'].apply(len)
filtered_df.head(50)
print(filtered_df['length_text'].max())
print(filtered_df['length_text'].min())
filtered_df['length_text'] = filtered_df['length_text'].astype('int')
```

```python
fig = px.histogram(filtered_df,
    x=filtered_df.index,
    y='length_text',
    nbins=100,
    color='label',
    template='plotly_white')
fig.update_layout(
    title={'text': "Lengths of the (Cleaned of Punctuation, Numbers, Stopwords, etc.) Text Opinions",
    'x': 0.5,
    'xanchor': 'center'})
fig.update_yaxes(title='Text Lengths')
fig.update_xaxes(showticksuffix="none")
fig.show()
```
